```python
from pathlib import Path
from utils.merge_game_summaries import merge_game_summaries
import pickle
from datetime import datetime
import matplotlib.pyplot as plt
from datetime import datetime
from collections import Counter

merged_file = merge_game_summaries(
    summary_dir=Path("../game_summaries"),
    merged_summary_dir=Path("merged_summaries"),
    used_summary_dir=Path("used_summaries"),
    base_filename="BetterNet_summary"
)
```

```python
with open("merged_summaries/BetterNet_summary_1.pkl","rb") as f:
    summaries = pickle.load(f)
    
summaries.sort(key=lambda s: datetime.fromisoformat(s["finished_at"]))
```

```python
num_turns = [s['num_turns'] for s in summaries]
plt.figure()
plt.hist(num_turns, bins=20)
plt.title('Distribution of Game Lengths (# Turns)')
plt.xlabel('Number of Turns')
plt.ylabel('Count')
plt.show()
```

```python
wins = sum(1 for s in summaries if s['winner'] == s['player'])
losses = len(summaries) - wins
plt.figure()
plt.bar(['Win', 'Loss'], [wins, losses])
plt.title('Overall Win vs Loss')
plt.ylabel('Count')
plt.show()
```

```python
all_patrons = [p for s in summaries for p in s['chosen_patrons']]
patron_counts = Counter(all_patrons)
patron_ids = sorted(patron_counts.keys())
counts = [patron_counts[p] for p in patron_ids]
plt.figure()
plt.bar([str(p) for p in patron_ids], counts)
plt.title('Patron Selection Frequency')
plt.xlabel('Patron ID')
plt.ylabel('Count')
plt.show()
```

```python
timestamps = [datetime.fromisoformat(s['finished_at']) for s in summaries]
avg_moves = [s['avg_moves_per_turn'] for s in summaries]
# Sort by time
time_sorted, moves_sorted = zip(*sorted(zip(timestamps, avg_moves), key=lambda x: x[0]))
plt.figure()
plt.plot(time_sorted, moves_sorted)
plt.title('Avg Moves per Turn Over Time')
plt.xlabel('Game Finish Time')
plt.ylabel('Avg Moves per Turn')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
