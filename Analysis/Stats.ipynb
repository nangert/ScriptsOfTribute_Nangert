```python
from pathlib import Path
from utils.merge_game_summaries import merge_game_summaries
import pickle
from datetime import datetime
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
```

```python
merged_dir = Path("merged_summaries")
all_summaries = []

for pkl in merged_dir.glob("BetterNet_summary_*.pkl"):
    with open(pkl, "rb") as f:
        batch = pickle.load(f)      # batch is a list of dicts
    all_summaries.extend(batch)     # concatenate into one list

# 2) Sort entire list by timestamp
all_summaries.sort(key=lambda s: datetime.fromisoformat(s["finished_at"]))

# now you can treat `all_summaries` just like your previous `summaries`
print(f"Loaded {len(all_summaries)} games from {merged_dir}.")
```

```python
num_turns = [s['num_turns'] for s in all_summaries]
plt.figure()
plt.hist(num_turns, bins=20)
plt.title('Distribution of Game Lengths (# Turns)')
plt.xlabel('Number of Turns')
plt.ylabel('Count')
plt.show()
```

```python
wins = sum(1 for s in all_summaries if s['winner'] == s['player'])
losses = len(all_summaries) - wins
plt.figure()
plt.bar(['Win', 'Loss'], [wins, losses])
plt.title('Overall Win vs Loss')
plt.ylabel('Count')
plt.show()
```

```python
all_patrons = [p for s in all_summaries for p in s['chosen_patrons']]
patron_counts = Counter(all_patrons)
patron_ids = sorted(patron_counts.keys())
counts = [patron_counts[p] for p in patron_ids]
plt.figure()
plt.bar([str(p) for p in patron_ids], counts)
plt.title('Patron Selection Frequency')
plt.xlabel('Patron ID')
plt.ylabel('Count')
plt.show()
```

```python
pairs = sorted(
    ((datetime.fromisoformat(s["finished_at"]), s["avg_moves_per_turn"]))
    for s in all_summaries
)

_, moves_sorted = zip(*pairs)

# 3) plot vs. episode index
plt.figure()
plt.plot(range(1, len(moves_sorted)+1), moves_sorted, marker=".")
plt.title("Avg Moves per Turn Over Chronological Episodes")
plt.xlabel("Episode Index (chronological)")
plt.ylabel("Avg Moves per Turn")
plt.tight_layout()
plt.show()
```

```python
window_size = 100  # change this to smooth more/less
# compute moving average via convolution
weights = np.ones(window_size) / window_size
rolling_avg = np.convolve(moves_sorted, weights, mode='valid')

# Episode indices for the rolling avg (centering it if you like)
# Here it starts at episode index = window_size
x = np.arange(window_size, len(moves_sorted) + 1)

plt.figure(figsize=(8,4))
plt.plot(moves_sorted, alpha=0.3, label="raw")
plt.plot(x, rolling_avg, color="C1", label=f"{window_size}-episode MA")
plt.title("Avg Moves per Turn (raw + rolling average)")
plt.xlabel("Episode Index (chronological)")
plt.ylabel("Avg Moves per Turn")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
pair_stats = {}  # (p1,p2) → {'games': int, 'wins': int}
for s in all_summaries:
    # sorted tuple so (3,5) == (5,3)
    p1, p2 = sorted(s['chosen_patrons'])
    key = (p1, p2)
    won = (s['winner'] == s['player'])
    pair_stats.setdefault(key, {'games': 0, 'wins': 0})
    pair_stats[key]['games'] += 1
    if won:
        pair_stats[key]['wins'] += 1

# filter and prepare for plotting
min_games = 5
pairs, rates, counts = [], [], []
for (p1, p2), stats in pair_stats.items():
    if stats['games'] >= min_games:
        pairs.append(f"{p1}-{p2}")
        rates.append(stats['wins'] / stats['games'])
        counts.append(stats['games'])

# sort by win rate descending
order = np.argsort(rates)[::-1]
pairs = [pairs[i] for i in order]
rates = [rates[i] for i in order]
counts = [counts[i] for i in order]

plt.figure(figsize=(10,4))
plt.bar(pairs, rates)
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Win Rate")
plt.xlabel("Patron Pair (ID-ID)")
plt.title(f"Patron-Pair Win Rates (≥{min_games} games)")
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

```python
ind_stats = {}  # p → {'games': int, 'wins': int}
for s in all_summaries:
    won = (s['winner'] == s['player'])
    for p in s['chosen_patrons']:
        ind_stats.setdefault(p, {'games': 0, 'wins': 0})
        ind_stats[p]['games'] += 1
        if won:
            ind_stats[p]['wins'] += 1

patrons = sorted(ind_stats.keys())
ind_rates = [ind_stats[p]['wins'] / ind_stats[p]['games'] for p in patrons]
ind_counts = [ind_stats[p]['games'] for p in patrons]

plt.figure(figsize=(8,4))
plt.bar([str(p) for p in patrons], ind_rates)
plt.xlabel("Patron ID")
plt.ylabel("Win Rate")
plt.title("Individual Patron Win Rates")
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

```python
# raw number of turns per game, in chronological order
num_turns = [s["num_turns"] for s in all_summaries]

# choose your smoothing window (e.g. 50 games)
window_size = 50
weights = np.ones(window_size) / window_size

# compute the rolling average via 1D convolution
rolling_turns = np.convolve(num_turns, weights, mode="valid")

# x-axis indices for the rolling avg (it starts at game index = window_size)
x = np.arange(window_size, len(num_turns) + 1)

plt.figure(figsize=(8,4))
# raw noisy signal faintly in the background
plt.plot(num_turns, alpha=0.3, label="Raw #Turns per Game")
# smoothed curve on top
plt.plot(x, rolling_turns, color="C2", label=f"{window_size}-game Rolling Avg")
plt.title("Rolling Average of Game Length (# Turns)")
plt.xlabel("Episode Index (chronological)")
plt.ylabel("Number of Turns")
plt.legend()
plt.tight_layout()
plt.show()
```
