```python
pair_stats = {}  # (p1,p2) → {'games': int, 'wins': int}
for s in all_summaries:
    # sorted tuple so (3,5) == (5,3)
    p1, p2 = sorted(s['chosen_patrons'])
    key = (p1, p2)
    won = (s['winner'] == s['player'])
    pair_stats.setdefault(key, {'games': 0, 'wins': 0})
    pair_stats[key]['games'] += 1
    if won:
        pair_stats[key]['wins'] += 1

# filter and prepare for plotting
min_games = 5
pairs, rates, counts = [], [], []
for (p1, p2), stats in pair_stats.items():
    if stats['games'] >= min_games:
        pairs.append(f"{p1}-{p2}")
        rates.append(stats['wins'] / stats['games'])
        counts.append(stats['games'])

# sort by win rate descending
order = np.argsort(rates)[::-1]
pairs = [pairs[i] for i in order]
rates = [rates[i] for i in order]
counts = [counts[i] for i in order]

plt.figure(figsize=(10,4))
plt.bar(pairs, rates)
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Win Rate")
plt.xlabel("Patron Pair (ID-ID)")
plt.title(f"Patron-Pair Win Rates (≥{min_games} games)")
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# assume `all_summaries` is already your chronological list
win_flags = np.array([1 if s['winner'] == s['player'] else 0 for s in all_summaries])
models   = [s['model'] for s in all_summaries]

# rolling window:
window_size = 50
weights = np.ones(window_size) / window_size
rolling_winrate = np.convolve(win_flags, weights, mode='valid')
x = np.arange(window_size, len(win_flags) + 1)

# find the episode‐indices where model changes
boundaries = []
current = models[0]
for i, m in enumerate(models):
    if m != current:
        boundaries.append((i, m))
        current = m

plt.figure(figsize=(10,4))
# raw wins (dots)
plt.plot(win_flags, '.', alpha=0.15, label="Raw Wins=1 / Loss=0")
# smoothed
plt.plot(x, rolling_winrate, lw=2, label=f"{window_size}-Game Rolling Win Rate")

# draw verticals and label
for idx, model_name in boundaries:
    plt.axvline(idx, color='grey', ls='--', alpha=0.7)
    plt.text(idx+1, 
             1.02,           # just above the top of the plot
             model_name, 
             rotation=90, 
             va='bottom', 
             fontsize=8, 
             alpha=0.8)

plt.title("Rolling Win Rate Over Episodes (with Model‐Switch Markers)")
plt.xlabel("Episode Index (chronological)")
plt.ylabel("Win Rate")
plt.ylim(-0.05, 1.05)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python

model_winrates = []
start = 0
current_model = models[0]

for idx, next_model in boundaries:
    # compute win‐rate for [start, idx)
    segment_wins = win_flags[start:idx].sum()
    segment_total = idx - start
    model_winrates.append((current_model,
                           segment_wins / segment_total))

    # advance
    start = idx
    current_model = next_model

# last segment (from last boundary to end)
segment_wins = win_flags[start:].sum()
segment_total = len(win_flags) - start
model_winrates.append((current_model,
                       segment_wins / segment_total))

print(model_winrates)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Split stats by role for individual patrons ---
roles = ["PLAYER1", "PLAYER2"]

ind_stats_by_role = {r: {} for r in roles}
for s in all_summaries:
    role = s["player"]                      # 'PLAYER1' or 'PLAYER2'
    won = (s["winner"] == s["player"])      # agent won?
    for p in s["chosen_patrons"]:
        d = ind_stats_by_role[role].setdefault(p, {"games": 0, "wins": 0})
        d["games"] += 1
        if won:
            d["wins"] += 1

# unify patron set and compute rates per role
patron_ids = sorted(set().union(*[set(ind_stats_by_role[r].keys()) for r in roles]))
rates_p1 = []
rates_p2 = []
for p in patron_ids:
    g1 = ind_stats_by_role["PLAYER1"].get(p, {"games": 0, "wins": 0})
    g2 = ind_stats_by_role["PLAYER2"].get(p, {"games": 0, "wins": 0})
    r1 = (g1["wins"] / g1["games"]) if g1["games"] else np.nan
    r2 = (g2["wins"] / g2["games"]) if g2["games"] else np.nan
    rates_p1.append(r1)
    rates_p2.append(r2)

# Plot grouped bars for patrons
x = np.arange(len(patron_ids))
w = 0.4
plt.figure(figsize=(10, 4))
plt.bar(x - w/2, rates_p1, width=w, label="PLAYER1")
plt.bar(x + w/2, rates_p2, width=w, label="PLAYER2")
plt.xticks(x, [str(p) for p in patron_ids], rotation=0)
plt.xlabel("Patron ID")
plt.ylabel("Win Rate")
plt.title("Individual Patron Win Rates by Player Seat")
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()


# --- Split stats by role for patron pairs ---
pair_stats_by_role = {r: {} for r in roles}
for s in all_summaries:
    role = s["player"]
    won = (s["winner"] == s["player"])
    p1, p2 = sorted(s["chosen_patrons"])
    key = (p1, p2)
    d = pair_stats_by_role[role].setdefault(key, {"games": 0, "wins": 0})
    d["games"] += 1
    if won:
        d["wins"] += 1

# filter, sort, and prepare _comp_data
min_games_total = 5  # keep pairs that appear at least this many times in total (P1+P2)
all_pairs = sorted(set().union(*[set(pair_stats_by_role[r].keys()) for r in roles]))

def total_games(k):
    return sum(pair_stats_by_role[r].get(k, {"games": 0})["games"] for r in roles)

def total_rate(k):
    g = total_games(k)
    if not g:
        return -1
    w = sum(pair_stats_by_role[r].get(k, {"wins": 0})["wins"] for r in roles)
    return w / g

kept_pairs = [k for k in all_pairs if total_games(k) >= min_games_total]
# sort by combined (P1+P2) win rate desc
kept_pairs.sort(key=lambda k: total_rate(k), reverse=True)

labels = [f"{a}-{b}" for (a, b) in kept_pairs]
rates_p1_pairs = []
rates_p2_pairs = []
for k in kept_pairs:
    g1 = pair_stats_by_role["PLAYER1"].get(k, {"games": 0, "wins": 0})
    g2 = pair_stats_by_role["PLAYER2"].get(k, {"games": 0, "wins": 0})
    r1 = (g1["wins"] / g1["games"]) if g1["games"] else np.nan
    r2 = (g2["wins"] / g2["games"]) if g2["games"] else np.nan
    rates_p1_pairs.append(r1)
    rates_p2_pairs.append(r2)

# Plot grouped bars for pairs
x = np.arange(len(kept_pairs))
w = 0.4
plt.figure(figsize=(12, 4))
plt.bar(x - w/2, rates_p1_pairs, width=w, label="PLAYER1")
plt.bar(x + w/2, rates_p2_pairs, width=w, label="PLAYER2")
plt.xticks(x, labels, rotation=90)
plt.ylabel("Win Rate")
plt.xlabel("Patron Pair (ID-ID)")
plt.title(f"Patron-Pair Win Rates by Player Seat (≥{min_games_total} total games)")
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from collections import defaultdict
import math

def as_enum(p):
    # normalize chosen_patrons entries to PatronId
    return p if isinstance(p, PatronId) else PatronId(int(p))

AVAILABLE_PATRONS = [
    PatronId.PELIN,
    PatronId.DUKE_OF_CROWS,
    PatronId.ANSEI,
    PatronId.RAJHIN,
    PatronId.ORGNUM,
    PatronId.SAINT_ALESSIA,
]

roles = ["PLAYER1", "PLAYER2"]

seat_stats = {r: defaultdict(lambda: {"games": 0, "wins": 0}) for r in roles}
overall_stats = defaultdict(lambda: {"games": 0, "wins": 0})

for s in all_summaries:
    role = s["player"]                   # 'PLAYER1' or 'PLAYER2'
    won  = (s["winner"] == s["player"])
    for raw in s["chosen_patrons"]:
        p = as_enum(raw)
        if p not in AVAILABLE_PATRONS:   # ignore patrons outside your 6-pack
            continue
        seat_stats[role][p]["games"] += 1
        overall_stats[p]["games"] += 1
        if won:
            seat_stats[role][p]["wins"] += 1
            overall_stats[p]["wins"] += 1

def safe_rate(w, g):
    return w / g if g else float("nan")

def sort_key_for_role(p, role):
    g_role = seat_stats[role][p]["games"]; w_role = seat_stats[role][p]["wins"]
    g_all  = overall_stats[p]["games"];    w_all  = overall_stats[p]["wins"]
    r_role = safe_rate(w_role, g_role)
    r_all  = safe_rate(w_all, g_all)
    # NaNs to -inf so they sink
    r_role_sort = r_role if not math.isnan(r_role) else float("-inf")
    r_all_sort  = r_all  if not math.isnan(r_all)  else float("-inf")
    # primary: seat-specific win rate, then seat sample size, then overall rate, then overall size
    return (r_role_sort, g_role, r_all_sort, g_all)

def preferred_order_for_role(role):
    return sorted(AVAILABLE_PATRONS, key=lambda p: sort_key_for_role(p, role), reverse=True)

PREFERRED_ORDER_P1 = preferred_order_for_role("PLAYER1")
PREFERRED_ORDER_P2 = preferred_order_for_role("PLAYER2")

# also as numbers
PREFERRED_ORDER_P1_NUMS = [p.value for p in PREFERRED_ORDER_P1]
PREFERRED_ORDER_P2_NUMS = [p.value for p in PREFERRED_ORDER_P2]

print("PREFERRED_ORDER_P1 =", PREFERRED_ORDER_P1)
print("PREFERRED_ORDER_P2 =", PREFERRED_ORDER_P2)
print("PREFERRED_ORDER_P1_NUMS =", PREFERRED_ORDER_P1_NUMS)
print("PREFERRED_ORDER_P2_NUMS =", PREFERRED_ORDER_P2_NUMS)
```
